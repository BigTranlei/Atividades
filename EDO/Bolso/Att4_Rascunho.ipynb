```python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# Parmetros do Problema
# ==============================

 # Comprimento x = 3, Comprimento y = 2, Potencial na borda x = 3

xx = np.linspace(0, 3, 200)
yy = np.linspace(0, 2, 200)
XX, YY = np.meshgrid(xx, yy)

def potencial(x, y):
    return np.sum(
        (4 * 20 * np.sinh(k * np.pi * x / 2 ) * np.sin(k * np.pi * y / 2 )) /
        (k * np.pi * np.sinh(k * np.pi * 3 / 2))
        for k in range(1, 150 ,2)
        )

Vmap = potencial(XX, YY)

# ==============================
# Plotagem
# ==============================

plt.contourf(XX, YY, Vmap, levels=55, cmap="inferno")
plt.colorbar(label="Potencial (V)")
plt.title("Mapa de Potencial no Guia de Onda")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
