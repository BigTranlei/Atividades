```python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# Parmetros do Problema
# ==============================

a = 3.0   # comprimento em x
b = 2.0   # comprimento em y
N = 150   # nmero de interao

xx = np.linspace(0, a, 200)
yy = np.linspace(0, b, 200)
XX, YY = np.meshgrid(xx, yy)

# ==============================
# Potenciais nas bordas
# ==============================

V1 = 5.0   # y=0 (inferior)

# ==============================
# Definies das quatro contribuies
# ==============================

def potencial_bottom(x, y, V=V1):
    # Borda em y=0
    return sum(
        (4*V * np.sinh(m*np.pi*(b-y)/a) * np.sin(m*np.pi*x/a)) /
        (m*np.pi * np.sinh(m*np.pi*b/a))
        for m in range(1, N, 2) # m mpar
    )

Vmap = potencial_bottom(XX, YY)

# ==============================
# Plotagem
# ==============================

plt.contourf(XX, YY, Vmap, levels=55, cmap="inferno")
plt.colorbar(label="Potencial (V)")
plt.title("Mapa de Potencial no Retngulo")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# Parmetros do Problema
# ==============================

a = 3.0   # comprimento em x
b = 2.0   # comprimento em y
N = 150   # nmero de interao

xx = np.linspace(0, a, 200)
yy = np.linspace(0, b, 200)
XX, YY = np.meshgrid(xx, yy)

# ==============================
# Potenciais nas bordas
# ==============================

V1 = 5.0   # y=0 (inferior)
V2 = 10.0   # x=a (direita)
V3 = 15.0   # y=b (superior)
V4 = 20.0   # x=0 (esquerda)

# ==============================
# Definies das quatro contribuies
# ==============================

def potencial_right(x, y, V=V2):
    # Borda em x=a
    return sum(
        (4*V * np.sinh(k*np.pi*x/b) * np.sin(k*np.pi*y/b)) /
        (k*np.pi * np.sinh(k*np.pi*a/b))
        for k in range(1, N, 2) # m mpar
    )

def potencial_left(x, y, V=V4):
    # Borda em x=0
    return sum(
        (4*V * np.sinh(k*np.pi*(a-x)/b) * np.sin(k*np.pi*y/b)) /
        (k*np.pi * np.sinh(k*np.pi*a/b))
        for k in range(1, N, 2) # m mpar
    )

def potencial_bottom(x, y, V=V1):
    # Borda em y=0
    return sum(
        (4*V * np.sinh(m*np.pi*(b-y)/a) * np.sin(m*np.pi*x/a)) /
        (m*np.pi * np.sinh(m*np.pi*b/a))
        for m in range(1, N, 2) # m mpar
    )

def potencial_top(x, y, V=V3):
    # Borda em y=b
    return sum(
        (4*V * np.sinh(m*np.pi*y/a) * np.sin(m*np.pi*x/a)) /
        (m*np.pi * np.sinh(m*np.pi*b/a))
        for m in range(1, N, 2) # m mpar
    )

# ==============================
# Soma de todas as contribuies
# ==============================

def potencial_total(x, y):
    return (potencial_right(x, y) +
            potencial_left(x, y) +
            potencial_bottom(x, y) +
            potencial_top(x, y))

Vmap = potencial_total(XX, YY)

# ==============================
# Plotagem
# ==============================

plt.contourf(XX, YY, Vmap, levels=55, cmap="inferno")
plt.colorbar(label="Potencial (V)")
plt.title("Mapa de Potencial no Retngulo")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
